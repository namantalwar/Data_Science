```python
fig,ax = plt.subplots(figsize=(16,8))
ax.scatter(sample['Sales'],sample['Profit'])
ax.set_xlabel('Sales')
ax.set_ylabel('Profit')
plt.show()
```

```python
sns.lineplot(x='Discount',y='Profit',label='Profit',data=sample)
plt.legend()
plt.show()
```

```python
sns.lineplot(x='Quantity',y='Profit',label='Profit',data=sample)
plt.legend()
plt.show()
```

```python
sample.groupby('Segment')[['Profit','Sales']].sum().plot.bar(color=['green','blue'],figsize=(8,5))
plt.ylabel('Profit/loss and sales')
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.title('Segment Wise Selection in Each Region')
sns.barplot(x='Region',y='Sales',data=sample, hue='Segment',order=sample['Region'].value_counts().index,palette='rocket')
plt.xlabel('Region',fontsize=15)
plt.show()
```

```python
sample.groupby('Region')[['Profit','Sales']].sum().plot.bar(color=['yellow','skyblue'],figsize=(8,5))
plt.ylabel('Profit/Loss and Sales')
plt.show()
```

```python
ps = sample.groupby('State')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color=['pink','red'],figsize=(15,8))
plt.title("Profit/Loss and Sales Across States")
plt.xlabel('States')
plt.ylabel('Profit/Loss and Sales')
plt.show()
```

```python
top_states= sample['State'].value_counts().nlargest(10)
top_states
```

```python
sample.groupby('Category')[['Profit','Sales']].sum().plot.bar(color=['yellow','green'],alpha=0.9,figsize=(8,5))
plt.ylabel('Profit/Loss and Sales')
plt.show()
```

   <h3>Obsevations:</h3>
    1. Technology and Office Supplies have high Profit
    2. Furniture has less Profit

```python
ps = sample.groupby('Sub-Category')[['Sales','Profit']].mean().sort_values(by='Sales',ascending=False)
ps[:].plot.bar(color = ['red','purple'],figsize = (15,8))
plt.title('Profit/Loss and Sales across states')
plt.xlabel('Sub-category')
plt.ylabel('Profit/Loss and Sales')
plt.show()
```
